# ATM System 

- class ATM
- functions
    - Check Balance
    - Deposite
    - Withdraw
    - Exit
  

```python
class ATM:
    def __init__(self,Balence):
        self.Balence = Balence
    def Check_Balence(self):
        print(f"The Total Balence is {self.Balence}")
    def Deposite(self,Amount):
        if(Amount>0):
            self.Balence += Amount
            print(f"Amount is Deposite Successfully. and Total Balence is {self.Balence}")
        else:
            print("Deposite is not possible please Enter Valid Amount")
    def Withdraw(self,Amount):
        if(Amount<0):
            print("plase Enter Valid Ammount.")
        elif(Amount>self.Balence):
            print("Not Sufficiant Amount in your Account. Try Again.")
        else:
            self.Balence -=Amount
            print(f"Amount is Withdraw Successfully.and Total Balence is {self.Balence} ")
```

```python
atm = ATM(90)
```

```python
atm.Check_Balence()
```

```python
atm.Deposite(10)
```

```python
atm.Withdraw(20)
```

```python
atm.Deposite(-90)
```

```python
atm.Withdraw(999)
```
